```python
import numpy as np
import matplotlib.pyplot as plt
import random

np.random.seed(2)
def unit_ball(p_s: list, num_points = 1000) -> None:
    fig, ax = plt.subplots()
    
    ax.set_aspect('equal')
    theta = np.linspace(0, 2 * np.pi, num_points)
    x = np.cos(theta) 
    y = np.sin(theta)
    for p in p_s:
        y_unsigned = (np.ones(len(x))**p - np.abs(x)**p)**(1/p) # solve for |y|
        yy = [y_unsigned[i] if e > 0 else y_unsigned[i] * -1 for i, e in enumerate(y)] # pm sign
    
        plt.plot(x, yy, label = f"p = {p}")
        
    plt.legend()

p_s = [0.5, 1,1.1, 2, 3, np.inf] # norm order, p

unit_ball(p_s)
```
