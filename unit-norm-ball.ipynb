```python
import numpy as np
import matplotlib.pyplot as plt
import random

np.random.seed(2)
def unit_ball(p_s: list, num_points = 1000) -> None:
    fig, ax = plt.subplots()
    
    # Create a meshgrid for plotting
    ax.set_aspect('equal')
    x = np.linspace(-1, 1, 1000)
    for p in p_s:
        y_unsigned = (np.ones(len(x))**p - np.abs(x)**p)**(1/p) # solve for y
        xx = np.append(x, x[::-1])
        yy = np.append(y_unsigned, -y_unsigned)
    
        plt.plot(xx,yy, label = f"p = {p}")
        
    plt.legend()
        
# Show the plot

p_s = [0.5, 1,1.1, 2, 3, np.inf] # norm order, p

unit_ball(p_s)
```
