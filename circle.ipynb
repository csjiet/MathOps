```python
import numpy as np
import matplotlib.pyplot as plt

def circle(num_points=10000):
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.set_aspect('equal')
    ax.axhline(0, color='black', lw=0.5)
    ax.axvline(0, color='black', lw=0.5)
    ax.set_xlim([-1.0, 1.0])
    ax.set_ylim([-1.0, 1.0])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Circle')

    '''
    Generate radian input 0 - 2pi (0 deg - 360 deg)
    '''
    theta = np.linspace(0, 2 * np.pi, num_points)
    '''
    pendulum: With the cos curve, we trace the output that it generates from the function 
    y = cos(z) generates, which is used as input, x in this plot: x = 1 -> 0 -> -1 -> 0 -> 1, 
    as the radian, z that we generated moves from left to the right.
    '''
    x = np.cos(theta) 
    '''
    pendulum: With the sin curve, we trace the output that it generates from the function 
    y = sin(z) generates, which is used as input, y in this plot: y = 0 -> 1 -> 0 -> -1 -> 0, 
    as the radian, z that we generated moves from the left to the right
    '''
    y = np.sin(theta) 

    '''
    This two interacting components --- x, y will simulate drawing a unit circle, where each 
    coordinate touches the radius of the unit circle.
    '''
    ax.plot(x,y)
    plt.show()

circle()
```
